# Linear Regression Model

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate some random data for demonstration
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 3 * X + np.random.randn(100, 1)*7

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
X_new = np.linspace(0, 10, 100).reshape(-1, 1)
y_pred = model.predict(X_new)

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

```python
# Make predictions
X_new = np.array([[5]])  # New data point
y_pred = model.predict(X_new)
# Print the predicted output
print("Predicted y:", y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

# Load the Diabetes dataset
diabetes = load_diabetes()

# Extract the feature and target variables
X = diabetes.data[:, 2].reshape(-1, 1)  # Use a single feature: body mass index (column index 2)
y = diabetes.target

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
X_new = np.linspace(np.min(X), np.max(X), 100).reshape(-1, 1)
y_pred = model.predict(X_new)

# Plot the data points and the regression line
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Body Mass Index')
plt.ylabel('Disease Progression')
plt.title('Linear Regression Example - Diabetes Dataset')
plt.legend()
plt.show()
```
