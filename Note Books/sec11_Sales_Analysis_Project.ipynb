<h1 style="text-align: justify; direction:rtl">   
 پروژه تحلیل داده‌های فروش
</h1>

---
<p style="text-align: justify; direction:rtl">   
در این پروژه، شما با استفاده از ابزارهای pandas و NumPy داده‌های فروش یک کسب‌وکار را تحلیل خواهید کرد. 
داده‌ها شامل اطلاعاتی مانند تاریخ سفارش، تاریخ ارسال، مشتریان، محصولات، فروش، تخفیف و سود هستند. هدف این پروژه، یادگیری روش‌های مختلف تحلیل داده‌ها و تفسیر نتایج است.  
ستون‌های موجود در داده‌ها عبارتند از:
</p>


1. **Row ID**: شماره‌ی یکتا برای هر رکورد.
2. **Order ID**: شناسه‌ی سفارش.
3. **Order Date**: تاریخ ثبت سفارش.
4. **Ship Date**: تاریخ ارسال سفارش.
5. **Ship Mode**: روش ارسال سفارش.
6. **Customer ID**: شناسه‌ی مشتری.
7. **Customer Name**: نام مشتری.
8. **Segment**: نوع مشتری (مصرف‌کننده، شرکتی، یا خانه‌دار).
9. **Country**: کشور.
10. **City**: شهر.
11. **State**: ایالت.
12. **Postal Code**: کد پستی.
13. **Region**: منطقه‌ی جغرافیایی.
14. **Product ID**: شناسه‌ی محصول.
15. **Category**: دسته‌بندی محصول.
16. **Sub-Category**: زیرمجموعه‌ی دسته‌بندی.
17. **Product Name**: نام محصول.
18. **Sales**: مبلغ فروش.
19. **Quantity**: تعداد محصولات.
20. **Discount**: درصد تخفیف.
21. **Profit**: سود حاصل از فروش.

سوالات
