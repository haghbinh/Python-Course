An important first distinction from lists is that array slices are views on the original array. This means that
the data is not copied, and any modifications to the view will be reflected in the source
array:

```python
x = np.arange(8)
x_slice = x[5:8]
x_slice[1] = 12345
x
```

```python
x_slice[:] = 64
x
```

```python
x = np.arange(8)
x_slice2 = x[5:8].copy()
x_slice2[1] = 12345
x
```

```python
x_slice2
```
