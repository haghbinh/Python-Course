<h2 dir="rtl" style="text-align: justify;">پایتون چیست؟</h2>

<p dir="rtl" style="text-align: justify;">
پایتون یک زبان برنامه‌نویسی ساده و قدرتمند است که برای ساخت برنامه‌های کاربردی و وب استفاده می‌شود. این زبان از نوع «شی‌گرا» و «سطح بالا» است، یعنی به کمک آن می‌توان خیلی راحت و سریع برنامه نوشت.
</p>

<p dir="rtl" style="text-align: justify;">
یکی از ویژگی‌های مهم پایتون، ساده بودن دستورهای آن است. همین موضوع باعث شده یادگیری آن برای مبتدی‌ها بسیار آسان باشد. کسانی که با زبان‌های دیگر هم کار کرده‌اند، معمولاً پایتون را خواناتر و قابل‌فهم‌تر می‌دانند. این خوانایی بالا کمک می‌کند اعضای یک تیم راحت‌تر با هم همکاری کنند.
</p>

<p dir="rtl" style="text-align: justify;">
در پایتون می‌توان از «ماژول‌ها» و «پکیج‌ها» استفاده کرد. این یعنی کدی که یک‌بار نوشته می‌شود را می‌توان بارها در پروژه‌های مختلف استفاده کرد. این ویژگی باعث می‌شود زمان کمتری برای نوشتن برنامه‌های جدید صرف شود.
</p>

<p dir="rtl" style="text-align: justify;">
اگر برنامه‌نویسی یک ماژول خاص بنویسد، دیگران هم می‌توانند از آن استفاده کنند. اضافه کردن (Import) یا خروجی گرفتن (Export) از این ماژول‌ها بسیار ساده است.
</p>

<p dir="rtl" style="text-align: justify;">
پایتون یک مزیت خیلی بزرگ دارد: همه ابزارهای اصلی آن رایگان هستند! شما هم می‌توانید مفسر و هم کتابخانه‌های آن را رایگان روی هر سیستم‌عاملی نصب و استفاده کنید. این یعنی نیازی به خرید ابزارهای گران نیست.
</p>

<p dir="rtl" style="text-align: justify;">
اگر حس می‌کنید مطالب بالا برایتان پیچیده بود، نگران نباشید. در ادامه، همه مفاهیم به‌صورت کامل و ساده توضیح داده می‌شود. فقط کافی است بدانید پایتون یک زبان ساده، رایگان و کاربردی است که با آن می‌توانید نرم‌افزارهای موبایل، وب و بسیاری برنامه‌های دیگر را بسازید. اگر وقت بگذارید، می‌توانید برنامه‌های خیلی خوبی با پایتون بنویسید.
</p>


<h2 dir="rtl" style="text-align: justify;">چرا پایتون محبوب است؟</h2>

<p dir="rtl" style="text-align: justify;">
پایتون یکی از محبوب‌ترین زبان‌های برنامه‌نویسی دنیاست. اما فقط به خاطر درآمد بالا نیست. دلایل زیادی باعث شده افراد زیادی سراغ یادگیری و استفاده از پایتون بروند.
</p>

<p dir="rtl" style="text-align: justify;">
<b>۱. شرکت‌های بزرگ دنیا از پایتون استفاده می‌کنند:</b> شرکت‌هایی مثل گوگل، یوتیوب، فیسبوک، ناسا، دراپ‌باکس، اینستاگرام و حتی IBM از پایتون در پروژه‌های خود استفاده می‌کنند. مثلاً گوگل از پایتون برای هوش مصنوعی و برنامه‌های ابری استفاده می‌کند. دراپ‌باکس هم بیشتر بخش‌های مختلف خودش را با پایتون ساخته است.
</p>

<b>۳. مناسب برای افراد مبتدی:</b> اگر تازه می‌خواهید برنامه‌نویسی یاد بگیرید، پایتون انتخاب فوق‌العاده‌ای است. چون دستوراتش ساده است و سریع می‌توان نتیجه گرفت. همین باعث می‌شود یادگیری برای شما جذاب‌تر شود.
</p>

<p dir="rtl" style="text-align: justify;">
<b>۴. پایتون تقریباً برای همه‌چیز کاربرد دارد:</b> با پایتون می‌توان برنامه‌های وب، اپلیکیشن‌های گرافیکی، تحلیل داده، اتوماسیون و خیلی چیزهای دیگر ساخت. یعنی اگر فقط پایتون بلد باشید، می‌توانید بیشتر پروژه‌ها را انجام دهید. این یکی از ویژگی‌های خاص این زبان است.
</p>


<h2 dir="rtl" style="text-align: justify;">ویژگی‌های پایتون</h2>

<p dir="rtl" style="text-align: justify;">
پایتون یک زبان برنامه‌نویسی پرقدرت است که می‌توان آن را با زبان‌هایی مثل پرل، روبی، پی‌اچ‌پی و جاوا مقایسه کرد. اما چیزی که پایتون را خاص می‌کند، سادگی و امکانات فراوان آن است. در ادامه، مهم‌ترین ویژگی‌های پایتون را با زبان ساده مرور می‌کنیم:
</p>

<ul dir="rtl" style="text-align: justify;">

  <li>
    <b>خوانایی بالا:</b> دستورهای پایتون ساده و شبیه زبان انسان هستند. برای مثال، اگر بخواهیم چیزی را چاپ کنیم، فقط کافی‌ست بنویسیم <code>print</code>. همین سادگی باعث می‌شود کدها راحت‌تر خوانده شوند.
  </li>

  <li>
    <b>آسان برای شروع:</b> پایتون زبانی ساده است و خیلی سریع می‌توان با آن برنامه نوشت. به همین دلیل برای ساخت نمونه‌های اولیه و پروژه‌های ساده بسیار مناسب است.
  </li>

  <li>
    <b>کتابخانه‌های آماده:</b> پایتون مجموعه بزرگی از کتابخانه‌های آماده دارد که کارهای مختلف مثل کار با فایل، جستجو، ساخت وب‌سایت و... را ساده می‌کنند. در بیشتر زبان‌ها باید این‌ها را خودتان بسازید.
  </li>

  <li>
    <b>حالت تعاملی:</b> شما می‌توانید کدهای کوتاه را در محیط تعاملی پایتون تست کنید. یک محیط برنامه‌نویسی ساده هم همراه با پایتون نصب می‌شود تا شروع کار برایتان راحت‌تر باشد.
  </li>

  <li>
    <b>قابل توسعه:</b> اگر نیاز به امکانات بیشتر دارید، می‌توانید با افزودن ماژول‌هایی که حتی با زبان‌هایی مثل C نوشته شده‌اند، پایتون را گسترش دهید.
  </li>

  <li>
    <b>قابل استفاده در برنامه‌های دیگر:</b> پایتون را می‌توان در دل برنامه‌های دیگر قرار داد تا کاربران آن برنامه‌ها بتوانند با پایتون کدنویسی کنند.
  </li>

  <li>
    <b>سازگاری با همه سیستم‌عامل‌ها:</b> پایتون روی ویندوز، لینوکس، مک و بیشتر سیستم‌ها قابل اجرا است و طوری طراحی شده که در همه جا تقریباً به یک شکل کار کند.
  </li>

  <li>
    <b>کاملاً رایگان و متن‌باز:</b> استفاده از پایتون هیچ هزینه‌ای ندارد. می‌توانید آن را آزادانه دانلود، ویرایش و استفاده کنید.
  </li>

</ul>


<h2 dir="rtl" style="text-align: justify;">راه‌اندازی محیط پایتون چگونه است؟</h2>
<p dir="rtl" style="text-align: justify;">یکی از مهم‌ترین موضوعات هنگام کار کردن با هر زبان برنامه‌نویسی «راه‌اندازی» (setup) یک «محیط توسعه» (Development Environment) برای آن است که امکان اجرای کدهای نوشته شده به آن زبان را می‌دهد. بدون این محیط توسعه، کاربر هرگز قادر به بررسی اینکه کدی که نوشته کار می‌کند یا نه و آیا دارای خطا است یا خیر، نخواهد بود.</p><div id="kaprila_blog_faradars_string-4"></div>
<p dir="rtl" style="text-align: justify;">برای زبان برنامه نویسی پایتون، به چیزی با عنوان «مفسر» (interpreter) نیز نیاز است که کد را تبدیل به آنچه می‌کند که برای کامپیوتر قابل خواندن و اجرا است (و در واقع کل برنامه نوشته شده را می‌سازد). برای تبدیل کد، ابتدا باید از «شِل پایتون» (Python shell) استفاده کرد که اغلب مفسر را از طریق چیزی که خط «بَنگ» (bang) نامیده می‌شود فراخوانی می‌کند.</p>
<p dir="rtl" style="text-align: justify;">همانطور که برای ساخت یک برنامه کاربری یا فایل دو راه وجود دارد، برای این کار نیز دو روش موجود است. می‌توان برنامه را با یک «ویرایش‌گر متن» (text editor) ساده مانند WordPad یا ++Notepad ساخت. همچنین، می‌توان برنامه را با استفاده از شل پایتون آماده کرد. مزایا و معایبی برای هر روش وجود دارد که در ادامه تشریح خواهند شد.</p>

<h2 dir="rtl" style="text-align: right;">IDE چیست؟</h2>

<p dir="rtl" style="text-align: justify;">
IDE مخفف عبارت <b>محیط توسعه یکپارچه</b> است. این ابزار نرم‌افزاری همه چیزهایی را که یک برنامه‌نویس برای نوشتن، اجرا، ویرایش و تست برنامه نیاز دارد، در یک محیط واحد در اختیار او قرار می‌دهد.
</p>

<p dir="rtl" style="text-align: justify;">
به زبان ساده، اگر بخواهیم برنامه‌ای بسازیم، باید قسمت‌های مختلفی مثل نوشتن کد، طراحی ظاهر برنامه، بررسی خطاها و اجرای برنامه را انجام دهیم. با استفاده از IDE می‌توان همه این کارها را در یک برنامه انجام داد و لازم نیست برای هر مرحله، ابزار جداگانه‌ای نصب کنیم.
</p>

<p dir="rtl" style="text-align: justify;">
بسیاری از IDEها امکانات مفیدی مثل <b>کد نویسی سریع‌تر</b>، <b>کامل کردن خودکار کد</b>، <b>اشکال‌زدایی (Debugging)</b> و <b>نمایش خطاها</b> را دارند. این امکانات باعث می‌شوند نوشتن برنامه آسان‌تر و سریع‌تر شود.
</p>

<p dir="rtl" style="text-align: justify;">
IDEها معمولاً دارای بخش‌های زیر هستند:
</p>

<ul dir="rtl" style="text-align: justify;">
  <li>ویرایشگر کد (Code Editor)</li>
  <li>کامپایلر یا مفسر (برای اجرای کد)</li>
  <li>اشکال‌زدا (Debugger)</li>
  <li>ابزارهای تکمیل خودکار و پیشنهاد کد</li>
</ul>

<p dir="rtl" style="text-align: justify;">
برخی IDEها فقط برای یک زبان خاص ساخته شده‌اند. مثلاً PyCharm برای پایتون یا IntelliJ برای جاوا. ولی بعضی دیگر مثل VS Code یا Eclipse برای چند زبان قابل استفاده‌اند.
</p>

<p dir="rtl" style="text-align: justify;">
یکی از ویژگی‌های مهم IDEها این است که اگر در کد خطایی وجود داشته باشد، خیلی راحت آن را پیدا کرده و نشان می‌دهند. این کار به برنامه‌نویس کمک می‌کند سریع‌تر مشکلات را حل کند.
</p>

<p dir="rtl" style="text-align: justify;">
در پروژه‌های بزرگ و پیچیده، استفاده از IDE بسیار کاربردی‌تر از ویرایشگرهای ساده است. چون هم سرعت برنامه‌نویسی را بیشتر می‌کند و هم کدنویسی را مرتب و راحت‌تر می‌سازد.
</p>

<p dir="rtl" style="text-align: justify;">
در ادامه، تفاوت IDE و ویرایشگر متن را بررسی خواهیم کرد تا بدانیم کدام ابزار برای چه کاری مناسب‌تر است.
</p>


<h2 dir="rtl" style="text-align: right;">محبوب‌ترین محیط‌های توسعه (IDE) برای پایتون</h2>

<p dir="rtl" style="text-align: justify;">
برای برنامه‌نویسی با زبان پایتون، ابزارهای مختلفی وجود دارد که می‌توان از آن‌ها برای نوشتن، اجرا و تست کدها استفاده کرد. در این بخش، محبوب‌ترین IDEها و ویرایشگرهایی را که بیشتر برنامه‌نویسان پایتون استفاده می‌کنند معرفی می‌کنیم.
</p>

<ul dir="rtl" style="text-align: justify;">

  <li>
    <b>Jupyter Notebook:</b> محیطی تحت وب، ساده و تعاملی. مخصوص آموزش، تحلیل داده و یادگیری مرحله‌به‌مرحله. خروجی کد و نمودارها را همان‌جا نشان می‌دهد.
  </li>

  <li>
    <b>Google Colab:</b> نسخه آنلاین Jupyter که نیاز به نصب ندارد. روی فضای ابری اجرا می‌شود و برای دانشجویان بسیار مناسب است. فقط با داشتن حساب Gmail قابل استفاده است.
  </li>

  <li>
    <b>Visual Studio Code (VS Code):</b> ویرایشگر سبک و حرفه‌ای که با نصب افزونه Python، به یک IDE کامل تبدیل می‌شود. مناسب برای پروژه‌های بزرگ‌تر.
  </li>

  <li>
    <b>PyCharm:</b> یک IDE قدرتمند مخصوص پایتون از شرکت JetBrains. نسخه رایگان (Community) و حرفه‌ای (Professional) دارد. برای پروژه‌های حرفه‌ای مناسب است و امکانات زیادی برای دیباگ، تست و مدیریت پروژه‌ها فراهم می‌کند.
  </li>

  <li>
    <b>Thonny:</b> محیطی بسیار ساده و سبک، مخصوص مبتدی‌ها. اگر تازه‌کار هستید، Thonny بهترین گزینه برای شروع است.
  </li>

  <li>
    <b>Spyder:</b> مخصوص برنامه‌نویسان علمی و تحلیل‌گران داده. همراه Anaconda نصب می‌شود و برای کار با NumPy، Pandas و Matplotlib عالی است.
  </li>

</ul>

<p dir="rtl" style="text-align: justify;">
هر کدام از این محیط‌ها نقاط قوت خاص خود را دارند. اگر تازه شروع کرده‌اید، پیشنهاد می‌شود با <b>Jupyter Notebook</b> یا <b>Google Colab</b> شروع کنید و سپس بر اساس نیاز، به محیط‌های پیشرفته‌تر مثل VS Code یا PyCharm مهاجرت کنید.
</p>


<h2 dir="rtl" style="text-align: right;">Jupyter Notebook چیست؟</h2>

<p dir="rtl" style="text-align: justify;">
Jupyter Notebook یک ابزار بسیار پرکاربرد و محبوب برای نوشتن و اجرای کدهای پایتون است. با استفاده از این ابزار، شما می‌توانید کد بنویسید، آن را اجرا کنید، نتیجه‌اش را ببینید و حتی توضیحاتی به‌صورت متن، تصویر یا فرمول‌های ریاضی در کنار آن اضافه کنید.
</p>

<p dir="rtl" style="text-align: justify;">
Jupyter مخصوصاً برای <b>یادگیری برنامه‌نویسی</b>، <b>تحلیل داده‌ها</b> و <b>انجام پروژه‌های علمی و دانشگاهی</b> بسیار مناسب است. چون به شما اجازه می‌دهد مرحله به مرحله کد بنویسید، اجرا کنید و همان‌جا نتیجه را ببینید.
</p>

<p dir="rtl" style="text-align: justify;">

<h2 dir="rtl" style="text-align: right;">Visual Studio Code چیست؟</h2>

<p dir="rtl" style="text-align: justify;">
Visual Studio Code یا به اختصار <b>VS Code</b>، یک ویرایشگر کد بسیار محبوب و پرکاربرد است که توسط شرکت مایکروسافت ساخته شده است. این ابزار رایگان و سبک است و روی ویندوز، مک و لینوکس قابل نصب است.
</p>

<p dir="rtl" style="text-align: justify;">
VS Code از بیشتر زبان‌های برنامه‌نویسی مانند Python، JavaScript، C++، HTML و بسیاری دیگر پشتیبانی می‌کند. با نصب افزونه‌ها (Extensions) می‌توانید امکانات آن را برای هر زبان توسعه دهید.
</p>

<p dir="rtl" style="text-align: justify;">
در دوره آموزش پایتون، از VS Code به‌عنوان یکی از ابزارهای اصلی برای نوشتن، اجرای کد و ساخت پروژه‌های واقعی استفاده خواهیم کرد.
</p>

<p dir="rtl" style="text-align: justify;">
ویژگی‌های اصلی VS Code:
</p>

<ul dir="rtl" style="text-align: justify;">
  <li><b>سبک و سریع:</b> اجرای سریع و مناسب برای کامپیوترهای ضعیف‌تر</li>
  <li><b>تکمیل خودکار کد (Auto-complete):</b> پیشنهاد کدها هنگام نوشتن</li>
  <li><b>اشکال‌زدایی (Debugging):</b> امکان بررسی و پیدا کردن خطاها در کد</li>
  <li><b>امکان نصب افزونه‌ها:</b> برای اضافه کردن امکانات بیشتر مانند اجرای کد پایتون، کنترل نسخه (Git)، یا قالب‌بندی کد</li>
  <li><b>پشتیبانی از فارسی:</b> با نصب افزونه‌های لازم، منوی فارسی و کدنویسی راست‌چین نیز ممکن است</li>
</ul>

<p dir="rtl" style="text-align: justify;">
برای نوشتن برنامه‌های پایتون با VS Code، تنها کافی است:
</p>

<ol dir="rtl" style="text-align: justify;">
  <li>VS Code را از سایت رسمی <a href="https://code.visualstudio.com/">code.visualstudio.com</a> دانلود و نصب کنید.</li>
  <li>افزونه Python را از بخش Extensions نصب کنید.</li>
  <li>در صورت نیاز، Python را نیز روی سیستم خود نصب کنید.</li>
</ol>

<p dir="rtl" style="text-align: justify;">
اگر به دنبال محیطی ساده ولی حرفه‌ای برای برنامه‌نویسی هستید، VS Code یک انتخاب عالی است. با آن می‌توانید پروژه‌های واقعی پایتون را مدیریت و اجرا کنید.
</p>
