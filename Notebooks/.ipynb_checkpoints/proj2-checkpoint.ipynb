<h2 dir="rtl" style="text-align: justify;">پروژه پایانی پایتون متوسطه: مدیریت فروشگاه آنلاین کوچک</h2>

<p dir="rtl" style="text-align: justify;">
در این پروژه شما یک برنامه مدیریت فروشگاه آنلاین کوچک خواهید ساخت که قابلیت مدیریت محصولات، موجودی و سفارش‌ها را دارد.  
هدف از این پروژه تمرین استفاده از مفاهیم پایتون متوسطه شامل توابع پیشرفته، Lambda، map، filter، zip، Decorators، Closure، ماژول‌ها و مدیریت استثناها است.
</p>

<h2 dir="rtl" style="text-align: justify;">شرح پروژه</h2>

<p dir="rtl" style="text-align: justify;">
برنامه باید قابلیت‌های زیر را داشته باشد:
</p>

<ol dir="rtl" style="text-align: justify;">
<li>اضافه کردن محصول جدید با مشخصات (نام محصول، قیمت، تعداد موجودی).</li>
<li>مشاهده لیست محصولات و وضعیت موجودی هر محصول.</li>
<li>ثبت سفارش برای یک یا چند محصول و کاهش موجودی آن‌ها.</li>
<li>بازگرداندن محصول و افزایش موجودی.</li>
<li>استفاده از Decorator برای ثبت تراکنش‌ها (Log) و نمایش پیغام قبل و بعد عملیات.</li>
<li>استفاده از Closure برای نگهداری تعداد دفعات سفارش هر محصول.</li>
<li>استفاده از Lambda، map و filter برای جستجو و فیلتر کردن محصولات.</li>
<li>مدیریت خطاها شامل سفارش محصول ناموجود، وارد کردن تعداد منفی و ورودی‌های نامعتبر.</li>
<li>امکان ذخیره‌سازی داده‌ها در فایل برای استفاده مجدد در اجرای بعدی برنامه.</li>
</ol>

<h2 dir="rtl" style="text-align: justify;">پیشنهاد ساختار فایل‌ها و ماژول‌ها</h2>

<p dir="rtl" style="text-align: justify;">
می‌توانید پروژه را به چند فایل ماژول تقسیم کنید:
</p>

<ul dir="rtl" style="text-align: justify;">
<li><b>products.py</b> : شامل کلاس Product و توابع مدیریت محصولات</li>
<li><b>orders.py</b> : شامل کلاس Order و توابع مدیریت سفارش‌ها</li>
<li><b>transactions.py</b> : شامل Decorator برای ثبت تراکنش‌ها و گزارش عملیات</li>
<li><b>main.py</b> : برنامه اصلی که رابط کاربری متنی دارد و همه ماژول‌ها را وارد می‌کند</li>
</ul>

<h2 dir="rtl" style="text-align: justify;">نکات مهم در پیاده‌سازی</h2>

<p dir="rtl" style="text-align: justify;">
- برای جستجو و فیلتر محصولات از <code>filter</code> و <code>lambda</code> استفاده کنید.<br>
- برای به‌روزرسانی موجودی محصولات از <code>map</code> و <code>zip</code> بهره ببرید.<br>
- Decoratorها را برای ثبت تراکنش‌ها و نمایش پیغام قبل و بعد از عملیات به کار ببرید.<br>
- Closureها را برای نگهداری تعداد دفعات سفارش هر محصول استفاده کنید.<br>
- مدیریت خطاها شامل سفارش محصول ناموجود، تعداد منفی یا ورودی نامعتبر باشد.<br>
- ذخیره و بازیابی داده‌ها با فایل باعث می‌شود فروشگاه دائماً قابل استفاده باشد.
</p>
