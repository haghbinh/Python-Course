<h2 dir="rtl" style="text-align: justify;">پروژه عملی: مدیریت کتابخانه کوچک</h2>

<p dir="rtl" style="text-align: justify;">
در این پروژه، شما یک برنامه برای مدیریت یک کتابخانه کوچک خواهید ساخت.  
هدف این است که اطلاعات کتاب‌ها و اعضا را ذخیره کنید، جستجو و گزارش‌گیری انجام دهید و عملیات مختلف را روی داده‌ها اعمال کنید.  
این پروژه شامل استفاده از لیست‌ها، Tupleها، مجموعه‌ها، دیکشنری‌ها، رشته‌ها و ورودی/خروجی است.
</p>

<h2 dir="rtl" style="text-align: justify;">۱. تعریف داده‌ها</h2>

<p dir="rtl" style="text-align: justify;">
- هر کتاب با یک Tuple شامل (عنوان، نویسنده، سال انتشار) نمایش داده می‌شود.  
- کل کتاب‌ها در یک لیست ذخیره می‌شوند.  
- ژانرها و برچسب‌های هر کتاب در یک مجموعه (Set) ذخیره می‌شوند.  
- اطلاعات اعضا در یک دیکشنری با کلید نام عضو و مقدار لیستی از کتاب‌های امانتی ذخیره می‌شود.
</p>

<h2 dir="rtl" style="text-align: justify;">۲. عملیات مورد نیاز</h2>

<p dir="rtl" style="text-align: justify;">
برنامه باید بتواند عملیات زیر را انجام دهد:
</p>

<ul dir="rtl" style="text-align: justify;">
<li>افزودن کتاب جدید به لیست کتاب‌ها.</li>
<li>جستجوی کتاب بر اساس عنوان یا نویسنده.</li>
<li>مشاهده تمامی کتاب‌ها با جزئیات.</li>
<li>اضافه کردن ژانر یا برچسب جدید به کتاب‌ها.</li>
<li>اضافه کردن عضو جدید و ثبت امانت کتاب برای اعضا.</li>
<li>گزارش‌گیری از تعداد کتاب‌ها، اعضا و ژانرهای موجود.</li>
<li>ذخیره و بارگذاری اطلاعات از فایل متنی (text file) با استفاده از ورودی/خروجی.</li>
</ul>

<h2 dir="rtl" style="text-align: justify;">۳. نکات تکمیلی</h2>

<p dir="rtl" style="text-align: justify;">
- هنگام دریافت ورودی از کاربر از تابع `input()` استفاده کنید.  
- برای نمایش اطلاعات از `print()` و فرمت‌بندی با f-string یا `format()` بهره ببرید.  
- در گزارش‌گیری می‌توانید از متدهای `len()`, `max()`, `min()` و بررسی وجود عناصر در مجموعه‌ها استفاده کنید.  
- استفاده از حلقه‌ها و شرط‌ها الزامی است تا عملیات اضافه کردن، جستجو و گزارش‌گیری انجام شود.
</p>

<h2 dir="rtl" style="text-align: justify;">۴. تمرین تکمیلی</h2>

<p dir="rtl" style="text-align: justify;">
۱. برنامه‌ای بنویسید که کاربر بتواند کتاب جدید وارد کند و در لیست کتاب‌ها ذخیره شود.  
</p>
<p dir="rtl" style="text-align: justify;">
۲. برنامه‌ای بنویسید که کاربر بتواند کتاب‌ها را بر اساس عنوان یا نویسنده جستجو کند و اطلاعات کامل آن نمایش داده شود.  
</p>
<p dir="rtl" style="text-align: justify;">
۳. برنامه‌ای بنویسید که بتواند اعضای کتابخانه را مدیریت کند، کتاب‌های امانتی را اضافه یا حذف کند و گزارش بگیرد.  
</p>
<p dir="rtl" style="text-align: justify;">
۴. برنامه‌ای بنویسید که اطلاعات کتاب‌ها و اعضا را در یک فایل متنی ذخیره کند و بتواند دوباره بارگذاری نماید.  
</p>
<p dir="rtl" style="text-align: justify;">
۵. از مجموعه‌ها استفاده کنید تا ژانرهای تکراری حذف شوند و تعداد ژانرهای منحصربه‌فرد گزارش شود.  
</p>
<p dir="rtl" style="text-align: justify;">
۶. برنامه‌ای بنویسید که اطلاعات کتاب‌ها را مرتب و نمایش دهد و امکان فیلتر کردن بر اساس سال انتشار یا ژانر را فراهم کند.  
</p>
